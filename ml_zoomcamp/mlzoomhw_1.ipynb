```python
rolls = data[data["Make"] == "Rolls-Royce"]
```

```python
kk = rolls[["Engine HP", "Engine Cylinders", "highway MPG"]]
```

```python
k = kk.drop_duplicates()
```

```python
x = k.values
```

```python
XTX = x.T.dot(x)
```

```python
inv = np.linalg.pinv(XTX)
```

```python
inv.sum()
```
